1) What is data ingestion and why is it important in data science? Which Python library is commonly used?

ANS :

Data ingestion is the process of collecting and importing data from different sources such as CSV files, databases, sensors, or web services into a system for analysis.
It is important in data science because analysis and machine learning models require properly structured data to produce accurate results. Without correct ingestion, data cannot be organized or processed effectively. The most commonly used Python library for data ingestion and cleaning is Pandas, which provides powerful tools for loading, organizing and preparing datasets efficiently.



2) Why is data cleaning necessary before performing analysis or machine learning?

ANS:

Data cleaning is necessary because raw data often contains missing values, duplicate records, incorrect formats and inconsistent entries that can reduce the accuracy of analysis and machine learning models.
Cleaning the data improves its quality and reliability by removing errors and making it consistent and structured. This ensures better decision-making and increases the performance and accuracy of predictive models.



3) How do you load a CSV file into a Pandas DataFrame?

ANS:

A CSV file can be loaded into a Pandas DataFrame using the read_csv() function available in the Pandas library. This function reads the data from the CSV file and converts it into a structured table format called a DataFrame, where rows represent records and columns represent variables. This makes the dataset easier to view, manage and analyze.



4) What is the difference between head() and tail() functions in Pandas?

ANS:

The head() function is used to display the first few rows of a dataset, which helps users quickly understand the structure and contents of the data.

The tail() function is used to display the last few rows of the dataset. Both functions are useful for verifying whether the dataset has been loaded correctly and for previewing data before performing further analysis.



5) How can you identify missing values in a dataset?

ANS:

Missing values in a dataset can be identified using functions such as isnull(), notnull(), and info() available in the Pandas library. These functions help detect empty or undefined values in the dataset. Identifying missing values is an important step in data cleaning because it helps decide whether the missing data should be removed, replaced or modified before performing analysis or machine learning tasks.

```python
import pandas as pd
import numpy as np
data = {
'Name': ['Alice', 'Bob', 'Charlie', 'Alice', np.nan],
'Age': [25, 30, 200, 25, 22],
'City': ['New York', 'Los Angeles', np.nan, 'New York', 'Chicago'],
'Sales': ['100', '200', '150', '100', '300']
}
df = pd.DataFrame(data)
print("--- Original Data ---")
print(df)
df = df.drop_duplicates()
df['City'] = df['City'].fillna('Unknown')
df = df.dropna(subset=['Name'])
df['Sales'] = df['Sales'].astype(int)
median_age = df[df['Age'] < 100]['Age'].median()
df.loc[df['Age'] > 100, 'Age'] = median_age
print("\n--- Cleaned Data ---")
print(df)
print("\n--- Data Info ---")
df.info()
```
