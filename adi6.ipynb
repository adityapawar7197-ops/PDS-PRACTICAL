1. What is web scraping? Explain the role of the Requests library in web scraping.

ANS :

Web scraping is the process of automatically extracting data from websites using programs instead of manually copying information. It allows users to collect useful data such as text, images, tables and links from web pages for analysis or storage. The Requests library plays an important role in web scraping because it is used to send HTTP requests to websites and retrieve their HTML content. This content is then processed and analyzed using libraries like BeautifulSoup to extract required information.



2. What is BeautifulSoup? How does it help in parsing HTML?

ANS :     

BeautifulSoup is a Python library used for parsing and extracting data from HTML and XML documents. It helps organize webpage content into a structured format so that users can easily navigate and search elements such as tags, attributes and text. BeautifulSoup simplifies the process of locating specific information within web pages and makes data extraction efficient and reliable for web scraping tasks.




3. Differentiate between web scraping and web crawling.

ANS :     

Web scraping refers to extracting specific information from websites for analysis or storage, while web crawling refers to automatically browsing and indexing multiple web pages across the internet using automated programs called crawlers. Web scraping focuses on collecting targeted data from selected pages, whereas web crawling focuses on discovering and navigating links across many pages to gather information systematically.




4. What are the commonly used methods in BeautifulSoup?

ANS :    

BeautifulSoup provides several commonly used methods such as find() to locate the first matching HTML element, find_all() to locate multiple matching elements, get_text() to extract text content from tags, and select() to search elements using CSS selectors. These methods help users easily navigate webpage structure and extract required data efficiently during web scraping.




5. What are the ethical and legal considerations in web scraping?

ANS :

Ethical and legal considerations in web scraping include respecting website terms of service, avoiding scraping restricted or private data and not overloading servers with frequent requests. Users should always check whether scraping is allowed by reviewing the website’s robots.txt file and usage policies. It is also important to ensure that collected data is used responsibly and does not violate copyright or privacy rules. Following ethical guidelines ensures safe and responsible use of web scraping techniques.

```python
# Import required libraries
import requests
from bs4 import BeautifulSoup

# Step 1: Take URL input from user
url = input("Enter the URL to scrape: ")

# Step 2: Take HTML tag input
tag = input("Enter the HTML tag (e.g., h1, h2, p, a, div): ")

# Step 3: Optional class name
class_name = input("Enter class name (leave blank if not required): ")

# Step 4: Send HTTP request
try:
    response = requests.get(url)

    if response.status_code != 200:
        print("Failed to fetch webpage. Status code:", response.status_code)
        exit()

except Exception as e:
    print("Error while requesting webpage:", e)
    exit()

# Step 5: Parse HTML content
soup = BeautifulSoup(response.text, 'html.parser')

# Step 6: Extract elements
if class_name:
    elements = soup.find_all(tag, class_=class_name)
else:
    elements = soup.find_all(tag)

# Step 7: Display extracted data
if not elements:
    print("No elements found.")
else:
    print(f"\nFound {len(elements)} elements:\n")

for index, element in enumerate(elements, start=1):
    print(f"Element {index}:")
    print("Text:", element.get_text(strip=True))
    print("Attributes:", element.attrs)
    print("-" * 50)

# Step 8: Extract page title
title = soup.title.string if soup.title else "No title found"
print("\nPage Title:", title)


```
