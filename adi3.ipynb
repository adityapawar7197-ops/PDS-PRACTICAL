 1 . What is data wrangling? Difference between data wrangling and data
preprocessing?

ans.


Data wrangling is the process of cleaning, transforming, organizing and structuring
raw data into a suitable format for analysis. It involves handling missing values,
correcting inconsistencies, filtering unwanted data and converting data types. Data
preprocessing is a broader step mainly used before applying machine learning
models and includes tasks such as normalization, encoding categorical values and
scaling features. In simple terms, data wrangling prepares raw data for analysis,
while data preprocessing prepares cleaned data for modeling and prediction tasks.



2 . Importance of Exploratory Data Analysis (EDA) in Pandas?


ANS:


Exploratory Data Analysis (EDA) is important because it helps understand the
structure, quality and patterns present in the dataset before performing detailed
analysis or building models. Using Pandas, EDA allows users to inspect datasets,
summarize statistics, detect missing values, identify outliers and understand
relationships between variables. This process improves decision making and
ensures the dataset is suitable for further analysis, which increases the accuracy of
results.





3 . Differences between Series, DataFrame and Panel?


ANS:


A Series is a one-dimensional labeled data structure that stores a single column of
data with an index. A DataFrame is a two-dimensional data structure that stores
data in rows and columns like a table and is the most commonly used structure in
Pandas. A Panel is a three-dimensional data structure used to store multiple
DataFrames together, but it is now deprecated in modern Pandas versions and
replaced by other tools like multi-index DataFrames. These structures help organize
data based on complexity and analysis requirements.





4 . Use of isnull(), notnull(), dropna(), fillna()?


ANS:


The isnull() function is used to detect missing values in a dataset by identifying
empty entries. The notnull() function performs the opposite task by identifying
values that are present and valid in the dataset. The dropna() function removes
rows or columns that contain missing values, which helps clean incomplete data.
The fillna() function replaces missing values with suitable alternatives such as
mean, median, mode or user-defined values. These functions are essential for
maintaining data quality before analysis or modeling.








5 . What is column transformation and why is it important in data analysis?



ANS:



Column transformation is the process of modifying or converting column values
into a more useful format for analysis. It includes changing data types, scaling
values, encoding categorical data or creating new columns from existing ones.
Column transformation is important because raw data is often not in the correct
format required for analysis or machine learning. Proper transformation improves
data consistency, enhances interpretation and increases the performance of
analytical models, making results more reliable and meaningfu

```python
import pandas as pd
import numpy as np
data = {
    "Name": ["Amit", "Aditya", "Uday", "krishana", None],
    "Age": [20, 21, None, 19, 22],
    "Marks": [85, 90, 78, None, 88],
    "City": ["Pune", "Mumbai", "Pune", "Russia", "Mumbai"]
}
df = pd.DataFrame(data)
print("Original Data:\n", df)
# Check missing values
print("\nMissing Values:\n", df.isnull())
# Fill missing values
df["Age"] = df["Age"].fillna(df["Age"].mean())
df["Marks"] = df["Marks"].fillna(0)
df["Name"] = df["Name"].fillna("Unknown")
# Drop rows with any remaining null values
df = df.dropna()
# Change data type
df["Age"] = df["Age"].astype(int)
# Rename columns
df.rename(columns={"Marks": "Score"}, inplace=True)
# Add new column (Transformation)
df["Result"] = df["Score"].apply(lambda x: "Pass" if x >= 40 else "Fail")
# Filter data
filtered_df = df[df["Score"] > 80]
# Sort data
sorted_df = df.sort_values(by="Score", ascending=False)
# Grouping data
grouped = df.groupby("City")["Score"].mean()
# Display results
print("\nCleaned Data:\n", df)
print("\nFiltered Data (Score > 80):\n", filtered_df)
print("\nSorted Data:\n", sorted_df)
print("\nGrouped Data (Average Score by City):\n", grouped)
```
